```python
import pandas as pd
from textblob import TextBlob
```

```python
data = {
    "Review": [
        "This product is amazing",
        "I love this service",
        "It is okay",
        "Very bad experience",
        "Worst product ever",
        "Excellent quality",
        "Not good",
        "Average performance"
    ]
}

df = pd.DataFrame(data)
df
```

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["Review"].apply(get_sentiment)

df
```

```python
df["Sentiment"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Sentiment"].value_counts().plot(kind='bar')

plt.title("Sentiment Analysis Results")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

```python
df["Sentiment"].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Sentiment Distribution")
plt.ylabel("")

plt.show()
```

CONCLUSION

1. Positive reviews were the majority.

2. Negative reviews highlighted customer dissatisfaction.

3. Neutral reviews represented balanced opinions.

4. Sentiment analysis helps organizations understand customer feedback.

5. Businesses can use sentiment analysis to improve products and services.
